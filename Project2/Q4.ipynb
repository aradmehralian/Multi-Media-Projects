## Smoothing and Edge Filters

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def imshow(image , title , size = 10):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
aerial_veiew_img = cv2.imread("images/AerialView.jpeg")
imshow(aerial_veiew_img, "Aerial Veiew Image")
```

### Median Filter

```python
kernel_size = 5
median_filtered_img = cv2.medianBlur(aerial_veiew_img, kernel_size)
imshow(median_filtered_img, "Median Filtered Image")
```

### Gaussian Filter

```python
kernel_width, kernel_height = (5, 5)
gaussian_filtered_img = cv2.GaussianBlur(aerial_veiew_img, (kernel_width, kernel_height), 0)
imshow(gaussian_filtered_img, "Gaussian Filtered Image")
```

### Sharpening  Filter

```python
sharpening_kernel = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]])

sharpened_img = cv2.filter2D(aerial_veiew_img, -1, sharpening_kernel)
imshow(sharpened_img, "Sharpened Image")
```

### Sobel Edge Detection

```python
sobel_kernel_size = 5
sobel_img = cv2.Sobel(aerial_veiew_img, 0, 1, 0, sobel_kernel_size)
imshow(sobel_img, "Sobel Edge Detection")
```

### Cany Edge Detection

```python
canny_img1 = cv2.Canny(aerial_veiew_img, 50, 120)
imshow(canny_img1, "Canny edge Dtection (moderate treshold)")
```

Too many edges detcted. let's narrow our margins so that it finds less edges

```python
canny_img2 = cv2.Canny(aerial_veiew_img, 180, 210)
imshow(canny_img2, "Canny edge Dtection (narrow treshold)")
```

Much better edge detction

Now let's use auto canny edge detction

```python
blurred_img = cv2.blur(aerial_veiew_img, (5, 5))
median_val = np.median(blurred_img)
lower = int(max(0, 0.9 * median_val))
upper = int(min(255, 1.33 * median_val))

auto_canny_img = cv2.Canny(aerial_veiew_img, lower, upper)
imshow(auto_canny_img, "Auto Canny Edge Detection")
```
