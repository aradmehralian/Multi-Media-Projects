## Smoothing and Edge Filters

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def imshow(image , title , size = 10):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
aerial_veiew_img = cv2.imread("images/AerialView.jpeg")
imshow(aerial_veiew_img, "Aerial Veiew Image")
```
