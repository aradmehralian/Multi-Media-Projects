## Smoothing and Edge Filters

```python
from src.functions import imread, imshow

from skimage import feature, color
from skimage.filters import median, gaussian, unsharp_mask, sobel
```

```python
aerial_view_img = imread("images/AerialView.jpeg")
imshow(aerial_view_img, "Aerial view image", size=15)
```

### Median Filter

```python
median_filtered_img = median(aerial_view_img)
imshow(median_filtered_img, "median filtered image", size=15)
```

### Gaussian Filter

```python
gaussian_filtered_img = gaussian(aerial_view_img, sigma=0.25)
imshow(gaussian_filtered_img, "Gaussian filtered image", size=15)
```

### Sharpening  Filter

```python
sharpened_img = unsharp_mask(aerial_view_img)
imshow(sharpened_img, "Sharpened Image", size=15)
```

As we can see the **Sharpening filter** has the best performance

### Sobel Edge Detection

```python
gray_aerial_img = color.rgb2gray(aerial_view_img) # converting to gray for better edge detection
```

```python
sobel_img = sobel(gray_aerial_img) 
imshow(sobel_img, "Sobel Edge Detection", size=15)
```

### Cany Edge Detection

```python
canny_img1 = feature.canny(gray_aerial_img)
imshow(canny_img1, "Canny edge Dtection (moderate treshold)", size=15)
```

Too many edges detcted. let's narrow our margins so that it finds less edges

```python
canny_img2 = feature.canny(gray_aerial_img, low_threshold=0.31, high_threshold=0.315)
imshow(canny_img2, "Canny edge Dtection (narrow treshold)", size=15)
```

**Canny Edge Filter** is more customizable and has performed better.
