## Image Enhancement

```python
import numpy as np
from src.functions import imread, imshow
```

```python
image1 = imread('images/image1.jfif')
imshow(image1, "image1")
```

First let's find the data type of image matrix. This will come handy when we want to revert the output matrices to {0, 255} range

```python
print(image1.dtype)
```

Now that we have found image's data type let's normalize it.

```python
normalized_image1 = image1 / 255
```

### Linear transfrom

```python
transform1 = ((1-normalized_image1) * 255).astype(np.uint8) # reverting back to 0 to 255 scale.
imshow(transform1, "Linear Transform")
```

### Piecewise Linear Transform (1)

```python
piecewise_transformed1 = np.piecewise(
    normalized_image1,
    [normalized_image1 < 0.2, (normalized_image1 >= 0.2) & (normalized_image1 < 0.55), normalized_image1 >= 0.55],
    [lambda x: x, 0, lambda x:x]
)

piecewise_transformed1 = (piecewise_transformed1 * 255).astype(np.uint8)
imshow(piecewise_transformed1, "Piecewise transform (1)" )
```

### Piecewise Linear Transform (2)

```python
piecewise_transformed2 = np.piecewise(
    normalized_image1,
    [normalized_image1 < 0.4, (normalized_image1 >= 0.4) & (normalized_image1 < 0.55), normalized_image1 >= 0.55],
    [0, lambda x:x, 1]
)
piecewise_transformed2 = (piecewise_transformed2 * 255).astype(np.uint8)
imshow(piecewise_transformed2, "Piecewise transform (2)" )
```

### Piecewise Linear Transform (3)

```python
piecewise_transformed3 = np.piecewise(
    normalized_image1,
    [normalized_image1 < 0.4, (normalized_image1 >= 0.4) & (normalized_image1 < 0.55), normalized_image1 >= 0.55],
    [lambda x: 1.5 * x, 0.3, lambda x: 1.56 * x - 0.56]
)

piecewise_transformed3 = (piecewise_transformed3 * 255).astype(np.uint8)
imshow(piecewise_transformed3, "Piecewise transform (3)" )
```

### 2nd power

```python
image_transformed5 = (normalized_image1 ** 2)
image_transformed5 = (image_transformed5 * 255).astype(np.uint8)
imshow(image_transformed5, "square Power Transformation") 
```

### 2nd root

```python
image_transformed4 = np.sqrt(normalized_image1).round()
image_transformed4 = (image_transformed4 * 255).astype(np.uint8)
imshow(image_transformed4, "square Root Transformation") 
```
