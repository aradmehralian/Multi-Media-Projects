## Histograms

```python
import cv2
from matplotlib import pyplot as plt
from src.functions import * 
```

```python
trees = imread('images/trees.jpeg')
imshow(trees, "Trees")
```

```python
trees_B, trees_G, trees_R= cv2.split(trees)
```

```python
plt.figure(figsize=(15, 5))

# Blue channel histogram
plt.subplot(1, 3, 1)
plt.hist(trees_B.ravel(), bins=256, range=[0, 256], color='blue')
plt.title('Blue Channel Histogram')
plt.xlim([0, 256])

# Green channel histogram
plt.subplot(1, 3, 2)
plt.hist(trees_G.ravel(), bins=256, range=[0, 256], color='green')
plt.title('Green Channel Histogram')
plt.xlim([0, 256])

# Red channel histogram
plt.subplot(1, 3, 3)
plt.hist(trees_R.ravel(), bins=256, range=[0, 256], color='red')
plt.title('Red Channel Histogram')
plt.xlim([0, 256])

plt.tight_layout()
plt.show()
```

## Histogram Equalization

```python
abraham = imread('images/abraham.jpg')
```

In order to perform histogram equalization we can either convert the original RGB image to gray scale and then perform equalization, 
or we can convert the image to hsv space and perform equalization on the V (brightness) channel. The latter method is used.

```python
# Convert the image to HSV color space
hsv_image = cv2.cvtColor(abraham, cv2.COLOR_RGB2HSV)

# Apply histogram equalization to the V channel
hsv_image[:, :, 2] = cv2.equalizeHist(hsv_image[:, :, 2])

# Convert the image back to RGB color space
equalized_color_image = cv2.cvtColor(hsv_image, cv2.COLOR_HSV2RGB)
```

```python
# Display the original and equalized images along with their histograms
plt.figure(figsize=(20, 10))

# Original Color Image
plt.subplot(2, 4, 1)
plt.imshow(abraham)
plt.title('Original Color Image')
plt.axis('off')

# Equalized Color Image
plt.subplot(2, 4, 2)
plt.imshow(equalized_color_image)
plt.title('Equalized Color Image')
plt.axis('off')

# Histograms for Original Image
plt.subplot(2, 4, 3)
plot_histogram(abraham[:, :, 2], 'Original Blue Channel Histogram', 'blue')

plt.subplot(2, 4, 4)
plot_histogram(abraham[:, :, 1], 'Original Green Channel Histogram', 'green')

plt.subplot(2, 4, 5)
plot_histogram(abraham[:, :, 0], 'Original Red Channel Histogram', 'red')

# Histograms for Equalized Image
plt.subplot(2, 4, 6)
plot_histogram(equalized_color_image[:, :, 2], 'Equalized Blue Channel Histogram', 'blue')

plt.subplot(2, 4, 7)
plot_histogram(equalized_color_image[:, :, 1], 'Equalized Green Channel Histogram', 'green')

plt.subplot(2, 4, 8)
plot_histogram(equalized_color_image[:, :, 0], 'Equalized Red Channel Histogram', 'red')

plt.tight_layout()
plt.show()
```
